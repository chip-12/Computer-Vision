```python
import numpy as np
import matplotlib.pyplot as plt

# Define the size of the image
rows = 100
cols = 100

# Create an empty 2D array with zeros
image_array = np.zeros((rows, cols, 3), dtype=np.uint8)

#Choose two colours to assign
color2 = np.array([255,255,255])

# Fill the array with the specified pattern
for i in range(rows):
    for j in range(cols):
        if j <= rows - i - 1:
            image_array[i, j] = color2

# Print the resulting array
#print(image_array)
    

# Display the image array
plt.imshow(image_array, cmap='viridis', interpolation='nearest')
plt.show()
```

```python
# Define the size of the chessboard
board_size = 8

# Create an empty 3D array with zeros
chessboard = np.zeros((board_size, board_size, 3), dtype=np.uint8)

# Define colors for the chessboard
color1 = np.array([255, 255, 255])  # White
color2 = np.array([0, 0, 0])        # Black

# Fill the array with the chessboard pattern
for i in range(board_size):
    for j in range(board_size):
        if (i + j) % 2 == 1:
            chessboard[i, j, :] = color1
        else:
            chessboard[i, j, :] = color2

# Display the chessboard
plt.imshow(chessboard)
plt.xticks([])  # Hide x-axis ticks
plt.yticks([])  # Hide y-axis ticks
plt.show()
```

```python
import cv2
# Read the image using OpenCV
image_2 = '/Users/nandinidhiran/Downloads/two.jpeg'  
image_two = cv2.imread(image_2)
image_two_r = cv2.resize(image_two, (1000,750))

image_6 = '/Users/nandinidhiran/Downloads/six.png'
image_six = cv2.imread(image_6)
image_six_r = cv2.resize(image_six, (1000,750))

plt.figure(figsize = (10,10))

plt.subplot(1,2,1)
plt.imshow(image_two_r)
plt.axis('on') 
plt.title('two')

plt.subplot(1,2,2)
plt.imshow(image_six_r)
plt.axis('on')  
plt.title('six')
```

```python
def translate_image(image, dx, dy):

    rows, cols, channels = image.shape
    
    # Create an empty image for the translated result
    translated_image = np.zeros_like(image, dtype=np.uint8)
    
    # Apply translation manually
    for i in range(rows):
        for j in range(cols):
            new_i = i + dy
            new_j = j + dx
            
            # Check if the new indices are within bounds
            if 0 <= new_i < rows and 0 <= new_j < cols:
                translated_image[new_i, new_j] = image[i, j]
    
    return translated_image

# Define translation parameters (dx and dy)
dx = -350
dy = 0

# Translate the image
translated_image = translate_image(image_two_r, dx, dy)

# Display the original and translated images side by side
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_two_r, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title('Translated Image')
plt.axis('off')

plt.show()
```
